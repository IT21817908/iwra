```python
import pandas as pd
```

```python
#load the datasets
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('moviesData.csv')
```

```python
# Merge DataFrames on the common column 'id' and 'movieId'
movie_ratings = pd.merge(movies, ratings, left_on='id', right_on='movieId')
movie_ratings
```

```python
# drop columns
movie_ratings = movie_ratings.drop(['id','genre','original_language','overview','popularity','release_date','vote_average','vote_count','timestamp'], axis =1)
movie_ratings
```

```python
# create a new data set
# for userId and the movies they rated
user_ratings = movie_ratings.pivot_table(index=['userId'], columns=['title'], values='rating')

user_ratings.head()
```

```python
# Remove movies which has less than 10 users who rated it. and fill remaining NaN with 0
user_ratings = user_ratings.dropna(thresh=10,axis=1).fillna(0)
user_ratings
```

```python
# build the similarity matrix - model
# use built in method to get the pearson correlation
item_similarity_df = user_ratings.corr(method='pearson')
item_similarity_df.head()
```

```python
# Make the recommendations to the user based on the model we created

def get_similar_movies(movie_name, user_rating):
    # if user has rated a movie bad those similar movies should go down the list
    similar_score = item_similarity_df[movie_name] * (user_rating - 2.5) # so if the rating is lower than 2.5, it will give more negative values and push them down the list
    similar_score = similar_score.sort_values(ascending=False)

    return  similar_score
```

```python
# Create a fake user
action_lover = [("2001: A Space Odyssey", 5), ("25th Hour", 1), ("2 Days in Paris", 1)]

#create empty data frame to collect all the similar movies that can suggest to the user
similar_movies = pd.DataFrame()

for movie, rating in action_lover:
    similar_movies = similar_movies.append(get_similar_movies(movie, rating), ignore_index=True)

similar_movies.head()
# sum all the values row wise and  make it in descending order
similar_movies.sum().sort_values(ascending=False)
```

```python
# Save the model
import pickle

#movies file
pickle.dump(user_ratings, open('user_ratings.pkl', 'wb'))

#similarity file
pickle.dump(item_similarity_df, open('item_similarity.pkl', 'wb'))
```
